```python
import pandas as pd
def recomendacao(df_cliente):
    banco_dados = pd.read_csv('banco_dados.csv')
    cluster = df_cliente.merge(banco_dados,on=['id']).groupby('CLUSTER').agg({'id':'count'}).reset_index()
    cluster['RANK'] = cluster['id'].rank(ascending=False)    
    cluster.rename(columns={'id':'TOTAL'},inplace=True)

    resultado = banco_dados.merge(cluster,on=['CLUSTER'])

#resultado[~resultado['id'].isin(df_cliente['id'])][['id','RANK']]

    return resultado[~resultado['id'].isin(df_cliente['id'])][['id','RANK']].sort_values(by='RANK',ascending=True)
```

```python
caminho_arquivo=r'..\estaticos_portfolio1.csv'
df_cliente = pd.read_csv(caminho_arquivo,index_col=0)
recomendacao(df_cliente)
```
